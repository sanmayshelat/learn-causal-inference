```python
import numpy as np
import pandas as pd
import statsmodels.formula.api as smf
import matplotlib.pyplot as plt
import linearmodels as lm
import pyfixest as pf
```

Following The Effect

```python
from causaldata import gapminder
gm = gapminder.load_pandas().data
```

```python
gm['logGDPpercap'] = np.log(gm['gdpPercap'])
gm[['logGDPpercap_within', 'lifeExp_within']] = gm.groupby('country')[['logGDPpercap','lifeExp']].transform(lambda x: x-x.mean())
gm.head()
```

```python
m1 = smf.ols(
    'lifeExp_within ~ logGDPpercap_within - 1',
    data = gm
).fit()
m1.summary()
```

```python
m2 = smf.ols(
    'lifeExp ~ logGDPpercap + C(country)',
    data = gm
).fit()
m2.summary()
```

```python
m3 = smf.ols(
    'lifeExp ~ logGDPpercap + C(country) + C(year)',
    data = gm
).fit()
m3.summary()
```

```python
lm.PanelOLS.from_formula(
    'lifeExp ~ logGDPpercap + EntityEffects + TimeEffects',
    data = gm.set_index(['country', 'year'])
).fit() # cov_type='clustered', cluster_entity=True
```

```python
pf.feols(
    'lifeExp ~ logGDPpercap | country + year',
    data = gm,
    vcov = 'iid' #{'CRV1': 'country'}
).summary()
```

```python
gm['logGDPpercap_mean'] = gm.groupby(['country'])['logGDPpercap'].transform('mean')
lm.RandomEffects.from_formula(
    'lifeExp ~ 1 + logGDPpercap_mean + logGDPpercap_within',
    data = gm.set_index(['country', 'year'])
).fit() # cov_type='clustered', cluster_entity=True
```
